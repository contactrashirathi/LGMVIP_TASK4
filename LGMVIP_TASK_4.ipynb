**Task : Image to Pencil Sketch with Python**

In this task, we are given a picture and we have to make a pencil sketch of it in python. With the help of cv2 library which helps in image processing, we change image to grey image, then blur it and finally to pencil sketch.

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

```python
img = cv2.imread("/content/task2 lgmvip img.png")
```

```python
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21,21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

```python
def dodgeV2(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dodgeV2(img_gray, img_smoothing)
cv2_imshow(final_img)
```

```python
cv2_imshow(img)
cv2_imshow(final_img)
```
